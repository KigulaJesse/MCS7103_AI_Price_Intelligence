```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
df = pd.read_csv("../02_data_cleaning/clean_data/cars_merged.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()

df.isnull().sum()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["price_ugx"], kde=True, bins=40)
plt.title("Distribution of Car Prices (UGX)")
plt.xlabel("Price (UGX)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["total_price_ugx"], kde=True, bins=40)
plt.title("Distribution of Total Prices (UGX)")
plt.xlabel("Total Price (UGX)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="source", y="price_ugx")
plt.title("Price Comparison by Source")
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
df["make"].value_counts().head(20).plot(kind="bar")
plt.title("Top 20 Car Makes")
plt.ylabel("Count")
plt.show()
```
